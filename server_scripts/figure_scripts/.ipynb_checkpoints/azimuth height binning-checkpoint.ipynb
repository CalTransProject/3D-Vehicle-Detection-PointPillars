# Notebook to make a figure showing height grid

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
# Set radius (arbitrary value)
r = 15

# Elevation angles in degrees (from 0 to 15 degrees in 5-degree increments)
elevation_angles_deg = np.arange(-31, 31, 5)  # 0, 5, 10, 15 degrees

# Azimuth angles in degrees (from -60 to 60 degrees in 5-degree increments)
azimuth_angles_deg = np.arange(-40, 40, 5)  # -60, -55, ..., 55, 60 degrees

# Azimuth angles in degrees (from -60 to 60 degrees in 5-degree increments)
light_azimuth_angles_deg = np.arange(-181, 180, 5)  # -60, -55, ..., 55, 60 degrees
```

```python
def set_axes_equal(ax):
    '''Make axes of 3D plot have equal scale so that spheres appear as spheres,
    cubes as cubes, etc..  This is one possible solution to Matplotlib's
    ax.set_aspect('equal') and ax.axis('equal') not working for 3D.'''
    
    x_limits = ax.get_xlim3d()
    y_limits = ax.get_ylim3d()
    z_limits = ax.get_zlim3d()

    x_range = abs(x_limits[1] - x_limits[0])
    x_middle = np.mean(x_limits)
    y_range = abs(y_limits[1] - y_limits[0])
    y_middle = np.mean(y_limits)
    z_range = abs(z_limits[1] - z_limits[0])
    z_middle = np.mean(z_limits)

    # The plot bounding box is a sphere in the sense of the infinity
    # norm, hence I call half the max range the 'box size'.
    box_size = max([x_range, y_range, z_range]) / 2.0

    ax.set_xlim3d([x_middle - box_size, x_middle + box_size])
    ax.set_ylim3d([y_middle - box_size, y_middle + box_size])
    ax.set_zlim3d([z_middle - box_size, z_middle + box_size])
```

```python
# Create a figure and a 3D axis
fig = plt.figure(figsize=(20, 15))
ax = fig.add_subplot(111, projection='3d')

z_adj = 10

# Plot lines of constant elevation (horizontal grid lines)
for elev_deg in elevation_angles_deg:
    elev_rad = np.deg2rad(elev_deg)
    # Convert elevation angle to polar angle (measured from the positive z-axis)
    phi_polar = (np.pi / 2) - elev_rad
    theta = np.deg2rad(azimuth_angles_deg)
    x = r * np.cos(theta) * np.sin(phi_polar)
    y = r * np.sin(theta) * np.sin(phi_polar)
    z = r * np.cos(phi_polar) * np.ones_like(theta) + z_adj
    ax.plot(x, y, z, color='blue', linestyle='-', linewidth=1)

# Plot lines of constant azimuth (vertical grid lines)
for azim_deg in azimuth_angles_deg:
    azim_rad = np.deg2rad(azim_deg)
    phi_elev = np.deg2rad(elevation_angles_deg)
    phi_polar = (np.pi / 2) - phi_elev
    theta = azim_rad * np.ones_like(phi_polar)
    x = r * np.cos(theta) * np.sin(phi_polar)
    y = r * np.sin(theta) * np.sin(phi_polar)
    z = r * np.cos(phi_polar) + z_adj
    ax.plot(x, y, z, color='red', linestyle='-', linewidth=1)
    
# SAME BUT LIGHTER FOR 360
# Plot lines of constant elevation (horizontal grid lines)
for elev_deg in elevation_angles_deg:
    elev_rad = np.deg2rad(elev_deg)
    # Convert elevation angle to polar angle (measured from the positive z-axis)
    phi_polar = (np.pi / 2) - elev_rad
    theta = np.deg2rad(light_azimuth_angles_deg)
    x = r * np.cos(theta) * np.sin(phi_polar)
    y = r * np.sin(theta) * np.sin(phi_polar)
    z = r * np.cos(phi_polar) * np.ones_like(theta) + z_adj
    ax.plot(x, y, z, color='grey', alpha=0.2, linestyle='-', linewidth=1)

# Plot lines of constant azimuth (vertical grid lines)
for azim_deg in light_azimuth_angles_deg:
    azim_rad = np.deg2rad(azim_deg)
    phi_elev = np.deg2rad(elevation_angles_deg)
    phi_polar = (np.pi / 2) - phi_elev
    theta = azim_rad * np.ones_like(phi_polar)
    x = r * np.cos(theta) * np.sin(phi_polar)
    y = r * np.sin(theta) * np.sin(phi_polar)
    z = r * np.cos(phi_polar) + z_adj
    ax.plot(x, y, z, color='grey', alpha=0.2, linestyle='-', linewidth=1)
    
ax.plot(0, 0, 10, 'D', markersize=15, color='black')
ax.text(0, 2, 10, 'Velodyne Sensor', 
        ha='right', 
        va='bottom',
        fontsize=16) 

x = 20
x2 = 0
y = 5
y2 = 0
z = 10
z2 = 10

ax.plot([x, x2], [y, y2], [z, z2], 
        linestyle=':', 
        color='black',
        linewidth=1.5)

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

legend_elements = [
    Line2D([0], [0], color='red', label='Azimuth bin boundaries'),
    Line2D([0], [0], color='blue', label='Elevation bin boundaries')
]

# Add the legend
ax.legend(handles=legend_elements, 
         loc='upper right',  # you can adjust this location
         bbox_to_anchor=(0.75, 0.7),  # adjust these numbers to move the legend
         framealpha=1.0,
         prop={'size': 16})

ax.set_axis_off()
# Set the aspect ratio to 'auto' for better visualization
# ax.set_box_aspect([np.ptp(a) for a in [x, y, z]])
set_axes_equal(ax)

# Optional: Set viewing angle for better perspective
ax.view_init(elev=20, azim=225)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
def plot_grid_front_view(num_azimuth=8, num_elevation=6, figsize=(10, 8)):
    """
    Creates a front view of the grid with color-coded azimuth and elevation lines.
    
    Args:
        num_azimuth (int): Number of azimuth divisions
        num_elevation (int): Number of elevation divisions
        figsize (tuple): Figure size in inches
    """
    # Create figure and axis
    fig, ax = plt.subplots(figsize=figsize)
    
    # Set up the grid ranges
    x = np.linspace(-1, 1, num_azimuth+1)
    y = np.linspace(-1, 1, num_elevation+1)
    
    # Plot vertical lines (azimuth)
    for i in x:
        ax.axvline(i, color='red', alpha=0.2, linestyle='-', label='Azimuth' if i == x[0] else "")
    
    # Plot horizontal lines (elevation)
    for i in y:
        ax.axhline(i, color='blue', alpha=0.2, linestyle='-', label='Elevation' if i == y[0] else "")
    
    # Set equal aspect ratio to maintain square cells
    ax.set_aspect('equal')
    
    # Add labels and title
    ax.set_xlabel('Azimuth bin boundaries', fontsize=16)
    ax.set_ylabel('Elevation bin boundaries', fontsize=16)
#     ax.set_title('Azimuth and Elevation Angle Bins')

    point_x = 0.43
    point_y = 0.23
    binned_point_x = 0.25
    binned_point_y = 0
    
    offset = 0.05

    ax.plot(point_x, point_y, '.', markersize=10, color='black')
    ax.text(point_x + offset, point_y + 0.03, 'Original point', 
            ha='left', 
            va='bottom',
            fontsize=16)
    
    ax.plot(binned_point_x, binned_point_y, '.', markersize=10, color='black')
    ax.text(binned_point_x - offset, binned_point_y, 'Lookup coordinates', 
            ha='right', 
            va='top',
            fontsize=16)

#     ax.plot([point_x, binned_point_x], [point_y, binned_point_y], 
#         linestyle=':', 
#         color='grey',
#         linewidth=1.5)
    ax.annotate(
        '', xy=(binned_point_x, binned_point_y), xytext=(point_x, point_y),
        arrowprops=dict(arrowstyle='->', color='black', linestyle='dashed')
    )
    
    # Add legend
    legend = ax.legend(fontsize=14) 
    
    # Set limits slightly larger than the grid
    ax.set_xlim(-1.1, 1.1)
    ax.set_ylim(-1.1, 1.1)
    
#     ax.set_axis_off()
    
    return fig, ax

# Create the basic plot
fig, ax = plot_grid_front_view()
fig.savefig('point_binning.png', dpi=300)
plt.show()
```
