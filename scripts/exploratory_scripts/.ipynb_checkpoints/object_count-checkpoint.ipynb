```python
import os
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
object_count = dict()
```

```python
DATA_DIR_ROOT = '../../data'
```

```python
def count_labels(label_path):
    with open(label_path, 'r') as file:
        for line in file:
            parts = line.strip().split()
            if parts[0] in object_count.keys():
                object_count[parts[0]] += 1
            else:
                object_count[parts[0]] = 1
```

```python
# Go into the ARCS data folder
labels_dir_path = Path(DATA_DIR_ROOT, 'labels')
# For each .txt file
for label_file in labels_dir_path.iterdir():
    if str(label_file).endswith('.txt'):
        # For each label in the file
        labels = count_labels(label_file)
    
```

```python
print(object_count)
```

```python
# Unpack keys and values
labels = list(object_count.keys())
counts = list(object_count.values())

fontsize = 24

# Create the bar chart
plt.figure(figsize=(16, 16))
plt.bar(labels, counts)
plt.xlabel('Vehicle Type', fontsize=fontsize + 6)
plt.ylabel('Count', fontsize=fontsize + 6)
# plt.title('Label Frequency for Zelzah and Plummer Dataset', fontsize=20)
plt.xticks(rotation=45, fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.tight_layout()

plt.savefig('zelzah_and_plummer_label_distribution.png', dpi=300)

# Display this masterpiece
plt.show()
```
