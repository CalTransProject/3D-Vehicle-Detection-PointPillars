```python
import numpy as np
import os
```

```python
velodyne_dir = 'velodyne_points'
labels_dir = 'labels'
labels_norm_dir = 'normalized_labels'
```

```python
def delete_file_if_exists(file_path):
    print('deleting file ' + file_path)
    
    try:
        os.remove(file_path)
    except Exception as e:
        print(f"\nFailed to process file {file_path}. Error: {e}")
        
        # print(f"File {file_path} not found, moving on.")
```

```python
def remove_frame(directory, file_id):
    print()
    point_file = file_id + '.bin'
    label_file = file_id + '.txt'
    
    # Create file paths
    point_file_path = os.path.join(directory, velodyne_dir, point_file)
    label_file_path = os.path.join(directory, labels_dir, label_file)
    norm_label_file_path = os.path.join(directory, labels_norm_dir, label_file)

    # Delete files
    delete_file_if_exists(point_file_path)
    delete_file_if_exists(label_file_path)
    delete_file_if_exists(norm_label_file_path)
```
